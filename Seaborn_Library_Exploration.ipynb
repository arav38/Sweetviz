```python
#Seaborn?
#Seaborn is a Python data visualization library based on matplotlib. 
#It provides a high-level interface for drawing attractive and informative statistical graphics.
#In my opinion, Seaborn is easier & faster to use (less code) Vs MatPlotLib
```

```python

# Packages / libraries
import os #provides functions for interacting with the operating system
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
marketdata =  pd.read_csv(r"/content/Marketing Raw Data.csv")
marketdata.head()
```

```python
#BAR PLOTS
```

```python
# Example 1 - Total Revenue by Month

ax = sns.barplot(x="Month_ID", y="Revenue", data=marketdata )
#Takes mean by default

marketdata[['Month_ID', 'Revenue']].groupby('Month_ID', as_index = False).agg({'Revenue':'mean'})

# Notes: 
#the lines signify the confidence interval
```

```python
# Example 2 - Total Revenue by Month - Remove the Confidence Interval
ax = sns.barplot(x="Month_ID", y="Revenue", data=marketdata, ci=False)
```

```python

# Example 3 - Total Revenue by Month - Remove the Confidence Interval - By Promo
ax = sns.barplot(x="Month_ID", y="Revenue", data=marketdata, ci=False, hue = 'Promo')
```

```python
# Example 4 - Total Revenue by Month - Remove the Confidence Interval - By Promo - Changing direction
ax = sns.barplot(x="Revenue", y="Year", ci=False, hue = 'Promo', orient = 'h', data=marketdata)
```

```python
# Example 5 - Total Revenue by Month - Remove the Confidence Interval - By Promo - Changing direction - Changing Colour
ax = sns.barplot(x="Revenue", y="Year", ci=False, hue = 'Promo', orient = 'h', data=marketdata, color="#1CB3B1")
```

```python
#Histograms - give distrubution of data 
```

```python
# Example 1 - Investigating the distribution of Revenue

x = marketdata['Revenue'].values

sns.distplot(x,color= 'blue')

#Right skewd distribution as the mean is to the right.Basically it says you large number of data concentration on right hand side.
#Basically technically your mode>mean>median
```

```python
# Example 2 - Investigating the distribution of Revenue, adding the mean
x = marketdata['Revenue'].values
sns.distplot(x,color = 'blue')

#calcualtion of mean

mean =  marketdata['Revenue'].mean()

#now plotting mean side distribution

plt.axvline(mean, 0,1, color = 'Blue')
```

```python

# Example 3 - Investigating the distribution of Visitors, adding the mean

x = marketdata['Visitors'].values

sns.distplot(x, color = 'red');

# Calculating the mean
mean = marketdata['Visitors'].mean()

#ploting the mean
plt.axvline(mean, 0,1, color = 'blue')
```

```python
#ScatterPlots
```

```python

# Example 1 - Relationship between Marketing Spend and Revenue

ax = sns.scatterplot(x="Marketing Spend", y="Revenue", data=marketdata, color = 'green')
```

```python
# Example 2 - Relationship between Marketing Spend and Revenue - changing color, hue & Style

ax = sns.scatterplot(x="Marketing Spend", y="Revenue", data=marketdata, color = 'green', hue = 'Promo', style = 'Promo')
```

```python

# Example 3 - Relationship between Marketing Spend and Revenue - changing color & hue - adding size

ax = sns.scatterplot(x="Marketing Spend", y="Revenue", data=marketdata, color = 'green', hue = 'Promo', size = 'Revenue',
                    sizes=(20, 200))
```

```python
# Example 4 - Relationship between Marketing Spend and Revenue - changing color & hue - adding size - by day

ax = sns.scatterplot(x="Visitors", y="Revenue", data=marketdata, color = 'green', hue = 'Day_Name', size = 'Revenue',
                    sizes=(20, 200))
```

```python
#Line Gragh
```

```python
#Example 1 - Simple 1 line graph
# Assuming we want to investigate the Revenue by Date

# Notes: error bands show the confidence interval

ax = sns.lineplot(x='Week_ID', y='Revenue', data = marketdata)
```

```python
# Example 2 - Adding Categories

# By Promo
ax = sns.lineplot(x='Week_ID', y='Revenue', hue = 'Promo', data = marketdata)
```

```python
# Example 3 - By Promo with style
ax = sns.lineplot(x='Week_ID', y='Revenue', hue = 'Promo', style = 'Promo', data = marketdata)
```

```python
#Example 4 - By Promo with style & Increase the size & Remove error bars

# increase the size
sns.set(rc={'figure.figsize':(12,10)})

ax = sns.lineplot(x='Week_ID', y='Revenue', hue = 'Promo', style = 'Promo', data = marketdata, ci=None)
```

```python
# Example 5 - By Promo with style & Increase the size & Remove error bars & adding markers & by month

ax = sns.lineplot(x='Month_ID', y='Revenue', hue = 'Promo', style = 'Promo', data = marketdata, ci=None,  markers=True)
```

```python
# Example 6 - By Day_Name with style & Increase the size & Remove error bars & adding markers & by month

ax = sns.lineplot(x='Month_ID', y='Revenue', hue = 'Day_Name', style = 'Promo', data = marketdata, ci=None,  markers=True)
```

```python
# Example 7 - By Year with style & Increase the size & Remove error bars & adding markers & by month

ax = sns.lineplot(x='Year', y='Revenue', hue = 'Day_Name', style = 'Promo', data = marketdata, ci=None,  markers=True)
```

```python
#lmPlots
```

```python
# Example 1 - Relationship between Marketing Spend and Revenue

ax = sns.lmplot(x="Marketing Spend", y="Revenue", data=marketdata, height=9)

# Notes:
# What is Linear Regression: It is a predictive statistical method for modelling the relationship between x (independent variable) & y (dependent V).
# How it works (cost function MSE): https://towardsdatascience.com/machine-learning-fundamentals-via-linear-regression-41a5d11f5220
```

```python
# Example 2 - Relationship between Marketing Spend and Revenue - changing color, hue & Style

ax = sns.lmplot(x="Marketing Spend", y="Revenue", data=marketdata, hue = 'Promo', ci= False, height=9, markers=["o", "x", "+"])
```

```python
# Example 3 - Relationship between Marketing Spend and Revenue - by column

ax = sns.lmplot(x="Marketing Spend", y="Revenue", data=marketdata, col = 'Promo', ci= False, height=5, 
                line_kws={'color': 'red'}, 
                scatter_kws={'color':'#ADC11E'})
```

```python
# Example 4 - Relationship between Marketing Spend and Revenue - by column - by day - add Jitter too

ax = sns.lmplot(x="Visitors", y="Revenue", data=marketdata, col = 'Day_Name', ci= False, height=4, 
                line_kws={'color': '#031722'}, 
                scatter_kws={'color':'#1E84C1'},
               col_wrap=3,
               x_jitter=.3)
```
